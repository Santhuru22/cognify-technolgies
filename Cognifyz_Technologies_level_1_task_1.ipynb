### *Check for Missing Values & Handle Missing Values*

```python
df.isnull().sum()
df.fillna(df.mean(numeric_only=True), inplace=True)
df.fillna("Unknown", inplace=True)
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
good_restaurants = df[df["Aggregate rating"] >= 4.0]
poor_restaurants = df[df["Aggregate rating"] < 3.0]
print(f"Number of Highly Rated Restaurants (4.0+): {len(good_restaurants)}")
print(f"Number of Poorly Rated Restaurants (<3.0): {len(poor_restaurants)}")
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Aggregate rating"], color="red", linewidth=2.5, width=0.4, fliersize=6, boxprops=dict(facecolor="lightblue"))
plt.xlabel("Aggregate Rating", fontsize=12, fontweight='bold', color='darkgreen')
plt.title("Box Plot of Restaurant Ratings", fontsize=14, fontweight='bold')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'], bins=18, kde=True, color='purple')
plt.xlabel("Aggregate Rating")
plt.ylabel("Count of Restaurants")
plt.title("Distribution of Restaurant Ratings")
plt.show()
```
