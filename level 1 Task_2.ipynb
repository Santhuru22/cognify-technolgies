### Summarize Dataset

```python
df = pd.read_csv('Dataset .csv')
df.head()
df.describe()
```

### Count Unique Values in Categorical Variables

```python
cat_cols = ["Country Code", "City", "Cuisines"]
for col in cat_cols:
    print(f"Unique values in {col}: {df[col].nunique()}")
    print(df[col].value_counts().head(10), "\n")
```

### Distribution of Categorical Variables

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, col in enumerate(cat_cols):
    sns.barplot(y=df[col].value_counts().head(10).index,
                x=df[col].value_counts().head(10).values,
                ax=axes[i], palette="viridis")
    axes[i].set_title(f"Top 10 {col} Distribution")
    axes[i].set_xlabel("Count")

plt.tight_layout()
plt.show()
```

### Top Cuisines and Cities With The Highest Number of Restaurants.


```python
top_cuisines = df["Cuisines"].value_counts().head(10)
print("Top 10 Cuisines:", top_cuisines)
top_cities = df["City"].value_counts().head(10)
print("Top 10 Cities with Highest Restaurants:", top_cities)
```
